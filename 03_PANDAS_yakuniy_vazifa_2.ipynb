## Dataset
#### Ushbu vazifada biz turli davlatlarning turli yillarda elektr tokini qanday manbaalardan olganini ko'rsatuvchi jadval bilan ishlaymiz.
Jadval ustunlari ko'mir, gaz, gidro va boshqa manbaalardan ko'rsatilgan yilda qancha __terravat/soat__ energiya olinganini ko'rsatadi.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df.columns = ['davlat','kod','yil','komir','gaz','gidro','yangilanuvchi','quyosh','neft','shamol','atom']
df.head()
```

### 1-VAZIFA. DF hajmini chiqaring

```python
print(df.shape)
```

### 2-VAZIFA. Jadvalda mavjud davlatlar ro'yxatini alohida chiqaring

```python
countries = df['davlat'].unique()
display(countries)
```

### 3-VAZIFA. DF dan faqat AQSh (`United States`) uchun ma'lumotlarni ajratib oling va yangi `dfus` nomli o'zgaruvchiga saqlang

```python
dfus = df[df['davlat'] == 'United States']
display(dfus.head())
```

### 4-VAZIFA. `dfus` jadvalining indeksini `yil` ustuni bilan o'zgartiring. DF dan `yil` ustunini tashlab yuboring.

```python
dfus.set_index('yil', inplace=True)
display(dfus.head())
```

### 5-VAZIFA. AQSh 2020-yilda jami qancha elektr energiyasi ishlab chiqqan?

```python
total_production_2020 = dfus.loc[2020, ['komir', 'gaz', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']].sum()
display(total_production_2020)
```

### 6-VAZIFA. 2010-2020-yillarda AQSH turli manbaalardan o'rtacha qancha elektr toki olganini chiqaring. Natijani kichikdan kattaga qarab tartiblang.

```python
avg_production_2010_2020 = dfus.loc[2010:2020, ['komir', 'gaz', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']].mean()
display(avg_production_2010_2020.sort_values())
```

### 7-VAZIFA. AQSH gidro elektr stansiyalardan eng kam va eng ko'p elektr energiyasi olgan yillarni toping.

```python
min_gidro_year = dfus['gidro'].idxmin()
max_gidro_year = dfus['gidro'].idxmax()

print(f"Eng kam gidro energiya olingan yil: {min_gidro_year}")
print(f"Eng ko'p gidro energiya olingan yil: {max_gidro_year}")
```

### 8-VAZIFA.
#### 8.1 Asl `df` dan 2019-yil ma'lumotlarini ajratib oling.
#### 8.2 Yangi df ga `df19` deb nom bering
#### 8.3 `df19` indeksini `davlat` ustuniga o'zgartiring hamda `davlat` va `yil` ustunlarini tashlab yuboring.
#### 8.4 `df19` dan `World` indeksini o'chirib tashlang

```python
df19 = df[df['yil'] == 2019].copy()
df19.set_index('davlat', inplace=True)
df19.drop(columns=['yil'], inplace=True)
df19.drop('World', inplace=True)
display(df19.head())
```

#### 8.5 `df19` jadvali uchun umumlashtiruvchi statistikani chiqaring

```python
display(df19.describe())
```

### 9-VAZIFA. 2019-yilda har bir davlat jami qancha elektr energiya ishlab chiqarganini DF ko'rinishida chiqaring.
*ishora. `Series` obyektini `DF` ga o'tkazing


```python
total_production_2019 = df19.drop(columns=['kod']).sum(axis=1)
total_production_2019_df = total_production_2019.to_frame(name='total_production')
display(total_production_2019_df.sort_values(by='total_production', ascending=False))
```

### 10-VAZIFA.
#### 10.1. Asl DF dan 2000-2020-yillar, `France` uchun `atom`, `komir`, `gaz`, `quyosh` manbaalarini alohida `dffr` jadvaliga ko'chirib oling
#### 10.2. `dffr` jadvalini yil ustuni bo'yicha tartiblang

```python
dffr = df[(df['davlat'] == 'France') & (df['yil'] >= 2000) & (df['yil'] <= 2020)][['yil', 'atom', 'komir', 'gaz', 'quyosh']].copy()
dffr.sort_values(by='yil', inplace=True)
display(dffr)
```

#### 10.3 `dffr`uchun Korrelyasiya jadvalini chiqaring.

```python
display(dffr.corr())
```
